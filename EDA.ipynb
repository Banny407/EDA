```python
categorical_featurs = [
     'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn',
       'calculate_TotalCharges', 'difference_Totalcharges'
]

for feature in categorical_featurs:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=feature,hue='Churn',palette='viridis')
    plt.title(f'Count of {feature} vs Churn')
    plt.xticks(rotation=45)
    plt.show()
```

```python
sns.histplot(data=df, x='MonthlyCharges',bins=25, hue='Churn', kde=True, palette='tab10')
plt.title('Monthly Charges Distribution by Churn')
plt.ylabel("No of Customers")
plt.show()
```

```python
numerical_features = ['tenure','MonthlyCharges','TotalCharges','difference_Totalcharges','calculate_TotalCharges']

for feature in numerical_features:
    plt.figure(figsize=(10,5))
    sns.histplot(x=feature, hue='Churn', data=df,kde=True,palette ='tab10')
    plt.title(f'{feature} Distribution by Churn', fontweight='bold')
    plt.xlabel(feature)
    plt.ylabel('Count of Customer')
    plt.show()
```

```python
 arpu = df ['MonthlyCharges'].mean()
 print (f"Avarage Revenue per User (ARPU): ${arpu:.2f}")

 total_revenue = df['TotalCharges'].sum()
 print(f"Total Revenue: ${total_revenue:.2f}")

 churned_customers = df[df['Churn']=='Churned']
 revenue_churn = churned_customers['MonthlyCharges'].sum()
 print(f"Revenue Churn: ${revenue_churn:.2f}")


avg_customers_lifespan = df['tenure'].mean()
clv=arpu*avg_customers_lifespan 
print(f"Customer Lifetime Value (CLV): ${clv:.2f}")
```
